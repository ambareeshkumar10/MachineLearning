```python
# Dropping Columns

features = df.drop(['Gender', 'Orders'], axis = 1)

features.head()
```

```python
# Scaling the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(features)

features = scaler.transform(features)

Feature_Scale = pd.DataFrame(features, columns = df.columns[0:36])

Feature_Scale.head()
```

### K_Means Model

```python
# Build a model with multiple K

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

n_clusters = [4,5,6,7,8]
# init['k-means++', 'random', 'ndarray', 'callable']


for i in n_clusters:
    cluster = KMeans(n_clusters = i, init='random', random_state = 0)
    predict = cluster.fit_predict(Feature_Scale)
    
    score = silhouette_score(Feature_Scale, predict, random_state = 0)
    print(f'for n_clusters {i}, silhouette score is {score}')
```
